```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
sents = [79548, 50000, 25000, 10000, 5000, 600]
fastalign = [0.304, 0.31, 0.332, 0.355, 0.371, 0.427]
eflomal = [0.148, 0.156, 0.176, 0.199, 0.222, 0.284]
simalign = [0.219 for _ in range(6)]
```

```python
plt.plot(sents, fastalign, label='fast_align')
plt.plot(sents, eflomal, label='eflomal')
plt.plot(sents, simalign, label='SimAlign')
plt.xlabel('Sentence pairs')
plt.ylabel('AER')
plt.legend()
plt.savefig('aer.png', dpi=300)
```

```python
fastalign_prec = [0.622, 0.62, 0.603, 0.581, 0.564, 0.515]
eflomal_prec = [0.827, 0.828, 0.812, 0.798, 0.776, 0.707]
simalign_prec = [0.795 for _ in range(6)]
```

```python
plt.plot(sents, fastalign_prec, label='fast_align')
plt.plot(sents, eflomal_prec, label='eflomal')
plt.plot(sents, simalign_prec, label='SimAlign')
plt.xlabel('Sentence pairs')
plt.ylabel('Precision')
plt.legend()
plt.savefig('precision.png', dpi=300)
```

```python
mbert = [0.266, 0.222, 0.219]
xlm_r_word = [0.457, 0.391, 0.342]
xlm_r_subword = [0.402, 0.369, 0.372]

labels = ['Argmax', 'Itermax', 'Match']
plt.xticks([0.12,1.12,2.12], labels)
plt.xlabel('Method')
plt.ylabel('AER')
width = 0.12


plt.bar(np.arange(len(mbert)), mbert, width=width, label='mBERT')
plt.bar(np.arange(len(xlm_r_word)) + width, xlm_r_word, width=width, label='XLM-R (word)')
plt.bar(np.arange(len(xlm_r_subword)) + 2*width, xlm_r_subword, width=width, label='XLM-R (subword)')
plt.legend()
plt.savefig('aer-comparison.png', dpi=300)
```

```python
df = pd.DataFrame({
    'Method': ['Argmax', 'Itermax', 'Match'],
    'mBERT': [0.266, 0.222, 0.219],
    'XLM-R (word)': [0.457, 0.391, 0.342],
    'XLM-R (subword)': [0.402, 0.369, 0.372]
})

df.plot(x="Method", y=["mBERT", "XLM-R (word)", "XLM-R (subword)"], kind="bar")
```
